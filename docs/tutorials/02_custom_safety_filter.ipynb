# Custom Safety Filter

This tutorial sketches how to implement the `SafetyFilter` ABC.

```python
from __future__ import annotations

import numpy as np
from numpy.typing import NDArray

from parallelcbf.api import BarrierState, MetricDict, SafetyFilter, SafetyFilterResult, SafetyState

ArrayF32 = NDArray[np.float32]

class ClipSafetyFilter(SafetyFilter[ArrayF32, ArrayF32]):
    def reset(self, *, seed: int | None = None) -> None:
        _ = seed

    def filter_action(self, observation: ArrayF32, nominal_action: ArrayF32, safety_state: SafetyState) -> SafetyFilterResult[ArrayF32]:
        _ = observation
        _ = safety_state
        safe = np.clip(nominal_action, -0.5, 0.5).astype(np.float32)
        return SafetyFilterResult(safe_action=safe, nominal_action=nominal_action, modified=bool(np.any(safe != nominal_action)), barrier_states={}, metrics={"clip_limit": 0.5})

    def barrier_state(self) -> dict[str, BarrierState]:
        return {}

    def metrics(self) -> MetricDict:
        return {"clip_limit": 0.5}

print(ClipSafetyFilter().metrics())
```
